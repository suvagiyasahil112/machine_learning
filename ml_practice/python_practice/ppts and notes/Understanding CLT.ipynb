### Case Study: Rolling Dice

**Objective**:To demonstrate the Central Limit Theorem by simulating the rolling of a dice multiple times and observing the distribution of the sample means.

**Background**:When a single dice is rolled, the outcome is uniformly distributed between 1 and 6. However, according to the CLT, if we repeatedly roll a dice multiple times and calculate the average of these rolls, the distribution of these averages should approach a normal distribution as the number of rolls (sample size) increases.

### Methodology:

1.  **Simulate Rolling a Dice**:
    
    *   Roll a single dice and observe the outcomes.
        
    *   Increase the number of dice rolls (sample size).
        
    *   Calculate the mean of these rolls.
        
    *   Repeat the process multiple times to generate a distribution of means.
        
2.  **Visualize the Results**:
    
    *   Plot the distribution of a single dice roll.
        
    *   Plot the distribution of sample means for different sample sizes.
        
3.  **Analyze the Results**:
    
    *   Compare the distributions and observe how they approach normality as sample size increases.
        

### Python Code Implementation

Let's implement this case study in Python:

```python
import numpy as np

import matplotlib.pyplot as plt



# Function to simulate rolling dice
def roll_dice(n_rolls):
    return np.random.randint(1, 7, size=n_rolls)



# Simulate rolling dice and plot the distribution
sample_sizes = [1, 10, 30, 50, 100, 250]

means = []

# Create plots
plt.figure(figsize=(12, 8))

for i, size in enumerate(sample_sizes, 1):
    # Roll dice and calculate means multiple times
    results = [np.mean(roll_dice(size)) for _ in range(10000)]
    means.append(results)
    
    # Create subplot for each sample size
    plt.subplot(2, 3, i)
    plt.hist(results, bins=30, alpha=0.7, color='blue')
    plt.title(f'Sample Size = {size}')
    plt.xlabel('Mean of dice rolls')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
roll_dice(250)
```

### Analysis

*   **Distribution of Single Dice Roll**: As expected, the histogram for a sample size of 1 should show a uniform distribution across numbers 1 to 6.
    
*   **Distribution of Sample Means**: As the sample size increases, the histograms for the means should begin to resemble a normal distribution. This is because the variance of the sampling distribution decreases with increasing sample size, concentrating more sample means around the true mean (which is 3.5 for a fair dice).
    

### Conclusion

This case study illustrates the Central Limit Theorem by showing how the distribution of sample means approaches a normal distribution as the sample size increases, despite the original distribution being uniform. This is a powerful result that underpins many statistical methods and concepts, particularly in hypothesis testing and confidence interval estimation.
