1.  **Generate the Dataset**:
    
    *   We create synthetic data for house sizes and prices with some added noise to simulate real-world data.
        
2.  **Visualize the Dataset**:
    
    *   We plot the data points to visualize the relationship between house size and price.
        
3.  **Implement Gradient Descent**:
    
    *   We initialize the weight and bias.
        
    *   We define the learning rate and number of iterations for the gradient descent algorithm.
        
    *   We iteratively update the weight and bias using the gradients of the cost function with respect to the parameters.
        
4.  **Train the Model**:
    
    *   We run the gradient descent algorithm for the specified number of iterations, updating the parameters to minimize the cost function.
        
5.  **Evaluate the Model**:
    
    *   We compute the mean squared error (MSE) and root mean squared error (RMSE) to evaluate the performance of the model.
        
6.  **Plot the Results**:
    
    *   We plot the regression line along with the original data points to visualize the fit of the model.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
house_sizes = 2.5 * np.random.randn(100) + 25  # House sizes in 1000s of square feet
house_prices = 250 * house_sizes + 5000 + 1000 * np.random.randn(100)  # House prices in $1000s


```

```python
# Display the first few rows of the dataset
print("House Sizes (in 1000s sq ft):", house_sizes[:5])
print("House Prices (in $1000s):", house_prices[:5])


```

```python
# Visualize the dataset
plt.scatter(house_sizes, house_prices, color='blue', label='Data points')
plt.xlabel('House Size (1000s sq ft)')
plt.ylabel('House Price ($1000s)')
plt.title('House Sizes vs. Prices')
plt.legend()
plt.show()
```

```python

# Gradient Descent Parameters
learning_rate = 0.00001
num_iterations = 260

# Initialize weights and bias
weight = np.random.randn()
bias = np.random.randn()

# Gradient Descent Algorithm
m = len(house_sizes)  # Number of data points
```

```python

for i in range(num_iterations):
    
    # Predictions
    predictions = weight * house_sizes + bias
    
    # Calculate the error
    error = predictions - house_prices
    
    # Compute gradients
    weight_gradient = (2/m) * np.dot(house_sizes, error)
    bias_gradient = (2/m) * np.sum(error)
    
    # Update weights and bias
    weight -= learning_rate * weight_gradient
    bias -= learning_rate * bias_gradient
    
    # Print progress every 100 iterations
    if i % 100 == 0:
        cost = (1/m) * np.sum(error ** 2)
        print(f"Iteration {i}: Cost {cost}, Weight {weight}, Bias {bias}")

```

```python
# Final parameters after gradient descent
print(f"Final Weight: {weight}, Final Bias: {bias}")

```

```python
# Predictions using the trained model
predicted_prices = weight * house_sizes + bias

```

```python
# Visualize the results
plt.scatter(house_sizes, house_prices, color='blue', label='Data points')
plt.plot(house_sizes, predicted_prices, color='red', label='Regression Line')
plt.xlabel('House Size (1000s sq ft)')
plt.ylabel('House Price ($1000s)')
plt.title('House Sizes vs. Prices with Regression Line')
plt.legend()
plt.show()

```

```python
# Evaluate the model
mse = np.mean((house_prices - predicted_prices) ** 2)
rmse = np.sqrt(mse)
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
```

#### Objective

The objective of linear regression is to find the best-fitting line through the data points that minimizes the difference between the predicted values and the actual values. This difference is quantified using a cost function, typically the Mean Squared Error (MSE).
